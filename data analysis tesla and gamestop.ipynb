```python
import yfinance as yf

# Télécharger les données boursières de Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Afficher les premières lignes
tesla_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table")
df = None 

for table in tables:
    if "Tesla Quarterly Revenue" in str(table):
        df = pd.read_html(str(table))[0]
        break

if df is not None:
    df.columns = ["Date", "Revenue"]
    
    df["Revenue"] = df["Revenue"].str.replace(r"[\$,]", "", regex=True).astype(float)
    df.head()
    # Sauvegarde des données
    tesla_data.to_csv("tesla_stock_data.csv")
else:
    print("Erreur : Tableau des revenus non trouvé. Vérifiez le site web.")
```

```python
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="max")

gamestop_data.head()

```

```python
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

tables = soup.find_all("table")
for table in tables:
    if "GameStop Quarterly Revenue" in str(table):
        df_gme = pd.read_html(str(table))[0]
        break

# Nettoyage des données
df_gme.columns = ["Date", "Revenue"]
df_gme["Revenue"] = df_gme["Revenue"].str.replace("[\$,]", "", regex=True).astype(float)

df_gme.head()
# Sauvegarde des données
df_gme.to_csv("gamestop_revenue_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# Charger les données
tesla_stock = pd.read_csv("tesla_stock_data.csv", index_col="Date", parse_dates=True)
tesla_revenue = pd.read_csv("tesla_revenue_data.csv")

# Créer la figure
fig, ax1 = plt.subplots(figsize=(12, 6))

# Tracer les données boursières
ax1.plot(tesla_stock.index, tesla_stock["Close"], color="blue", label="Tesla Stock Price")
ax1.set_ylabel("Stock Price (USD)", color="blue")
ax1.set_xlabel("Date")

# Ajouter un second axe pour les revenus
ax2 = ax1.twinx()
ax2.bar(tesla_revenue["Date"], tesla_revenue["Revenue"], alpha=0.5, color="red", label="Tesla Revenue")
ax2.set_ylabel("Revenue (Billion USD)", color="red")

plt.title("Tesla Stock Price and Revenue Over Time")
plt.legend()
plt.show()
```
